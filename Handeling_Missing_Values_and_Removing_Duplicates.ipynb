```python
import pandas as pd

# Suppose your data is in a CSV file
df = pd.read_excel('data_clean1.xlsx')
```

```python
import pandas as pd

# Replace NaN with average in numeric column
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        mean_value = df[column].mean()
        # Replace NaN with mean, ensuring mean_value is not NaN
        df[column].fillna(mean_value if pd.notna(mean_value) else 0, inplace=True)
    else:
        # Replace NaN in non-numeric columns with the appropriate value
        # For example, you can replace NaN with 'Unknown' or other strategies
        # Replace with values ​​based on certain categories or logic
        # Here is an example to replace with 'Unknown' or 'FALSE'/'TRUE'
        if df[column].dtype == 'object':  # If the column is an object type (string)

        # If the column only has True/False values, we can fill it with False
            if df[column].isin(['TRUE', 'FALSE']).all():
                df[column].fillna('FALSE', inplace=True)
            else:
        # For columns with categorical values, you can fill them with the most frequently occurring value (mode).                most_common = df[column].mode()[0]
                df[column].fillna(most_common, inplace=True)
        else:
        # If the column is not numeric and not an object, we can use 'NaN' or another default value.
            df[column].fillna('NaN', inplace=True)

# Display data after NaN replacement
df.head(42518)

# Save DataFrame to Excel file
df.to_excel('data_clean2.xlsx', index=False)
```
